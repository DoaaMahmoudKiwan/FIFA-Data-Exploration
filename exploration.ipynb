# Data Wrangling

```python
# Select some columns to be able to follow up
fifa_sub = fifa[['Age', "Nationality","Club",'Position', "Overall","Value","Wage","Preferred Foot","Height","Weight","Finishing","Penalties","Reactions","Stamina","BallControl","Dribbling"]].copy()
fifa_sub.info()
```

```python
#drop rows with empty entries
fifa_sub.dropna(axis = 0, inplace = True)
fifa_sub.reset_index(drop=True, inplace=True)
fifa_sub.info()
```

- Change the type of some variables to be manageble


```python
#Extract numbers from the Value column and convert them to floats
fifa_sub['Value'].replace({'[€]':'', '.5M':'500000', 'M':'000000', 'K':'000'}, regex=True, inplace=True)
fifa_sub["Value"]=pd.to_numeric(fifa_sub["Value"])/1000
#get rid of rows with value =0
fifa_sub= fifa_sub[fifa_sub["Value"]>0]
```

```python
#Extract numbers from the Wage column and convert them to floats
fifa_sub["Wage"] = fifa_sub["Wage"].apply(lambda x :x[1:-1])
indx = fifa_sub[fifa_sub["Wage"]==" "].index
fifa_sub.drop(fifa_sub.index[indx], inplace=True)
fifa_sub.reset_index(drop = True, inplace = True)
fifa_sub["Wage"]= pd.to_numeric(fifa_sub["Wage"])
```

```python
# Change the Weight to integer and then convert from pounds to kilos
fifa_sub["Weight"]= fifa_sub["Weight"].apply(lambda x: x[0:-3]).astype(int).apply(lambda x:  x/ 2.2046)
```

```python
# Change the Height to integer and then convert from feet to cms
#function to convert height to cm 
def feet_to_cm(col):
    temp = col.split("'")
    ft = int(temp[0])
    inch = int(temp[1])
    return ft * 0.3048 + inch * 0.0254

fifa_sub["Height"] =fifa_sub["Height"].apply(lambda x: feet_to_cm(x))
```

```python
def log_trans(x, inverse = False):
    """ quick function for computing log and power operations """
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)

fifa_sub['log_Value'] = fifa_sub['Value'].apply(log_trans)
fifa_sub['log_Wage'] = fifa_sub['Wage'].apply(log_trans)
```

## Multivariate Exploration


Now I want to investigate the relationship between players value , wage, position and preferred foot in two specific clubs (FC Barcelona and Juventus)

```python
# Create a dataframe with players of FC Barcelona and Juventus
df = fifa_sub[(fifa_sub["Club"]=="FC Barcelona")| (fifa_sub["Club"]=="Juventus")]
```

```python
position_order = df.Position.unique()
```

```python
def countplot(x, hue, **kwargs):
    data=kwargs.pop("data")
    sb.countplot(x, hue=hue, data=data,**kwargs)
    plt.ylabel("Count")
    plt.xlabel("Position")
g = sb.FacetGrid(df, row='Club',  height=4, aspect=2.5, sharex=False, sharey=False)
g.fig.suptitle("Number of Players in each Position with Diferrent Preferred Foot in FCB and Juventus", y=1.05)
g.map_dataframe(countplot, 'Position',hue='Preferred Foot',order = position_order,palette=sb.color_palette())
g.add_legend(title= "Preferred Foot");
```

The relation between the palyer position and preferred foot in Barcelona is different than that of Juventus. For instance, Barcelona has 4 leftfooted GoalKeepers (GK) and no rightfooted GoalKeepers while Juventus has 2 rightfooted GoolKeepers,....

```python
# Create facet scatter plots to show the relation between the value and wages of players of the two clubs
g = sb.FacetGrid(data = df, hue = 'Club', height = 5)
g.map(plt.scatter, 'Wage', 'log_Value')
g.add_legend()
plt.ylabel("Value (Euro)")
plt.title("Players Wage vs. Value for FC Barcelona and Juventus")
plt.ylim(log_trans(0,5))
plt.yticks(log_trans(np.array([1, 10, 100, 1e3, 1e4, 1e5])), ['1k', '10k', '100k', '1M','10M','100M']);
plt.xticks(np.arange(0,700,100), ['0', '100k', '200k', '300k','400k','500k','600k']);
```

The range of wages in Juventus is narrow when compared to that in Barcelona. Most wages in Juventus are <= 200k. However, in Barcelona, there is a wider range of wages. Therefore, players of the same market value recieve more wages in Barcelona than in Juventus. 

```python
plt.figure(figsize=(8,6))
ax= sb.barplot(data = df, x = 'Club', y = 'log_Value', hue = 'Preferred Foot')
ax.legend(ncol = 2, framealpha = 1, title = 'Preferred Foot')
ax.set_ylabel("Value (Euro)")
ax.title.set_text("Average Value of Players in FC Barcelona and Juventus for Different Preferred Foot")
ax.set_yticks(log_trans(np.array([1, 10, 100, 1e3, 1e4])))
ax.set_yticklabels(['1k', '10k', '100k', '1M','10M']);
#plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter('%d k'));
```

> Leftfooted palyers in Barcelona tend to have higher average value than Leftfooted palyers in Juventus.

> On the other hand, rightfooted palyers in Juventus tend to have higher average value than rightfooted palyers in Barcelona.

```python
# function to plot the heatmap of values between preferred foot and position of the palyer
def plot_heatmap_fun (club, i):
    t = df[df["Club"]== club]
    temp= t.groupby(['Preferred Foot', 'Position']).mean()['Value']
    temp= temp.reset_index(name = 'value_avg')
    temp= temp.pivot(index = 'Position', columns = 'Preferred Foot',values = 'value_avg')
    sb.heatmap(temp, annot = True, fmt = '.0f',cbar_kws = {'label' : 'mean(Value)/1000 (Euro)'}, ax=ax[i])
    ax[i].set_title(club)
    
    
```

```python
fig, ax = plt.subplots(ncols = 2, nrows = 1 , figsize = [24,24])
plot_heatmap_fun ("FC Barcelona", 0)
ax[0].title.set_text("Players Average Value for Different Poistions and Preferred Foot in FCB")
plot_heatmap_fun ("Juventus", 1)
ax[1].title.set_text("Players Average Value for Different Poistions and Preferred Foot in Juventus")
```

> The two clubs have very different distributions when it comes to the average value of leftfooted and rightfooted palyers in different positions.

> In FC Barcelona, the three highest average values are for leftfooted RFs, rightfooted RSs and rightfooted LWs.

>In Juventus, the three highest average values are for leftfooted LFs, rightfooted CDMs and Leftfooted LMs.
